```python
df.drop(['Order_ID'],axis=1,inplace=True)
```

```python
#data encoding
df['Traffic_Level']=df['Traffic_Level'].map({'Low':0,'Medium':1,'High':2})
df
```

```python
df=pd.get_dummies(df,columns=['Weather'],dtype='int')
df
```

```python
df=pd.get_dummies(df,columns=['Time_of_Day'],dtype='int')
df
```

```python
df=pd.get_dummies(df,columns=['Vehicle_Type'],dtype='int')
df
```

```python
#labeling data
x=df.drop('Delivery_Time_min',axis=1)
y=df['Delivery_Time_min']
x
```

```python
print(x.columns)
```

```python
#splitting data for train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#linear regression
from sklearn.linear_model import LinearRegression

#linear regression is a class ,o use class first we have to create a object

linear_regressor=LinearRegression()
#watever object name we gave it is a model(Linear_Regression)

#trainig model
linear_regressor.fit(x_train,y_train)
```

```python
#testing model
y_pred_lr=linear_regressor.predict(x_test)
y_pred_lr
```

```python
#model evaluation
#checking results based on y_test

print('R^2 score of LR on Training set',linear_regressor.score(x_train,y_train))
print('R^2 score of LR on Test set',linear_regressor.score(x_test,y_test))
```

```python
#checking for error (checking wrong prediction)
from sklearn.metrics import mean_absolute_error as mae,mean_squared_error as mse
print('--Linear Regression Model Evaluation--')
print('Mean Absolute Error',mae(y_test,y_pred_lr))
print('Mean squared Error',mse(y_test,y_pred_lr))
print('root mean squared error',np.sqrt(mse(y_test,y_pred_lr)))
                                             
```

```python

def FoodDeliveryTimePrediction():
    # Collect user inputs
    Distance_km = float(input("Enter distance (km): "))
    Traffic_Level = int(input("Enter traffic level (1=Low, 2=Medium, 3=High): "))
    Preparation_Time_min = int(input("Enter preparation time (minutes): "))
    Courier_Experience_yrs = int(input("Enter courier experience (years): "))
    
    # Weather (dummy encoding)
    weather = input("Enter weather (Clear/Foggy/Rainy/Snowy/Windy): ").strip().lower()
    Weather_Clear = 1 if weather == "clear" else 0
    Weather_Foggy = 1 if weather == "foggy" else 0
    Weather_Rainy = 1 if weather == "rainy" else 0
    Weather_Snowy = 1 if weather == "snowy" else 0
    Weather_Windy = 1 if weather == "windy" else 0
    
    # Time of Day (dummy encoding)
    time_of_day = input("Enter time of day (Morning/Afternoon/Evening/Night): ").strip().lower()
    Time_of_Day_Morning = 1 if time_of_day == "morning" else 0
    Time_of_Day_Afternoon = 1 if time_of_day == "afternoon" else 0
    Time_of_Day_Evening = 1 if time_of_day == "evening" else 0
    Time_of_Day_Night = 1 if time_of_day == "night" else 0
    
    # Vehicle Type (dummy encoding)
    vehicle = input("Enter vehicle type (Bike/Car/Scooter): ").strip().lower()
    Vehicle_Type_Bike = 1 if vehicle == "bike" else 0
    Vehicle_Type_Car = 1 if vehicle == "car" else 0
    Vehicle_Type_Scooter = 1 if vehicle == "scooter" else 0
    
    # Put into a dictionary with exact dataset column names
    order_features = {
        "Distance_km": Distance_km,
        "Traffic_Level": Traffic_Level,
        "Preparation_Time_min": Preparation_Time_min,
        "Courier_Experience_yrs": Courier_Experience_yrs,
        "Weather_Clear": Weather_Clear,
        "Weather_Foggy": Weather_Foggy,
        "Weather_Rainy": Weather_Rainy,
        "Weather_Snowy": Weather_Snowy,
        "Weather_Windy": Weather_Windy,
        "Time_of_Day_Afternoon": Time_of_Day_Afternoon,
        "Time_of_Day_Evening": Time_of_Day_Evening,
        "Time_of_Day_Morning": Time_of_Day_Morning,
        "Time_of_Day_Night": Time_of_Day_Night,
        "Vehicle_Type_Bike": Vehicle_Type_Bike,
        "Vehicle_Type_Car": Vehicle_Type_Car,
        "Vehicle_Type_Scooter": Vehicle_Type_Scooter
    }
    
    print("\nUser Input Features:")
    print(order_features)
    
    # Convert dictionary to DataFrame with one row
    order_df = pd.DataFrame([order_features])
    
    # Scale the input
    test_scaled = scaler.transform(order_df)
    
    # Predict
    prediction = rf.predict(test_scaled)
    print(f'Food Delivery Time Prediction: {prediction[0]:,.2f}')
    return prediction[0]

# Call the function
FoodDeliveryTimePrediction()
```
